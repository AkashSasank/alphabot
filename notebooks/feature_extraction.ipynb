# SBIN Feature Extraction (Since Inception)

This notebook mirrors the `sbin_feature_extract_example.py` flow:
1. Load environment variables
2. Build Kite session config
3. Initialize `BotManager` and `HistoricalFeatureExtractor`
4. Export SBIN daily features from inception to today as CSV

```python
from __future__ import annotations

import os
from datetime import datetime
from pathlib import Path

from dotenv import load_dotenv
from tradingbot.bot import BotManager
from tradingbot.feature_extractor import HistoricalFeatureExtractor


def build_session_config() -> dict[str, object]:
    """Build BotManager session config from environment variables."""
    return {
        "api_key": os.getenv("KITE_API_KEY"),
        "api_secret": os.getenv("KITE_API_SECRET"),
        "user_id": os.getenv("KITE_USER_ID"),
        "password": os.getenv("KITE_PASSWORD"),
        "pin": os.getenv("KITE_PIN"),
        "headless": os.getenv("KITE_HEADLESS", "False").lower() == "true",
        "timeout_ms": int(os.getenv("KITE_TIMEOUT_MS", "45000")),
        "redirect_url": os.getenv(
            "KITE_REDIRECT_URL",
            "http://localhost:1130/",
        ),
    }
```

```python
# Run SBIN extraction from inception through now
load_dotenv()

manager = BotManager(build_session_config())
extractor = HistoricalFeatureExtractor(manager)

# Using an early start date to capture full available history.
start_date = datetime(1996, 1, 1)
end_date = datetime.now()

output_path = Path("data/sbin_since_inception_day_features.csv")
output_path.parent.mkdir(parents=True, exist_ok=True)

print("Starting SBIN feature extraction...")
print(f"Date range: {start_date.isoformat()} -> {end_date.isoformat()}")

saved_path = extractor.export_features_csv(
    ticker_name="SBIN",
    interval="day",
    output_csv_path=str(output_path),
    start_date=start_date,
    end_date=end_date,
)

print(f"Extraction completed: {saved_path}")
```
